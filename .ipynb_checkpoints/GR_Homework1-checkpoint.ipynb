```julia
using Plots
using LaTeXStrings
using Printf
using Random

default(dpi=300)
```

```julia
function x(param)
    return param^2 - 1
end

function y(param)
    return (param - 1)/(param + 1)
end
```

```julia
# Define the range for param
param_range = -0.5:0.01:1.0

# Calculate values for both functions
x_values = x.(param_range)
y_values = y.(param_range)

# Create the plot
plot(param_range, x_values, label=L"x(\lambda) = \lambda^2 - 1", lw=2, color=:blue)
plot!(param_range, y_values, label=L"y(\lambda) = \frac{\lambda - 1}{\lambda + 1}", lw=2, color=:red)

# Add labels and title
xlabel!(L"\lambda")
ylabel!("Function Values")
```
