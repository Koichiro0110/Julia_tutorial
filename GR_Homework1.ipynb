```julia
using Plots
using LaTeXStrings
using Printf
using Random

default(dpi=300)
```

```julia
function x(param)
    return param^2 - 1
end

function y(param)
    return (param - 1)/(param + 1)
end
```

```julia
# Define the range for param
param_range = -0.5:0.01:1.0

# Calculate values for both functions
x_values = x.(param_range)
y_values = y.(param_range)

# Create the plot
plot(x_values, y_values, label=L"y = y(x(\lambda)), \lambda \in \left[-\frac{1}{2}, 1\right]", legendfontsize=12,
xtickfont=font(10, "Times New Roman"), ytickfont=font(10, "Times New Roman"),
    tick_direction=:out, framestyle=:box, linewidth=2, color=:blue)

# Add labels and title
xlabel!(L"x", fontsize=30)
ylabel!(L"y", fontsize=30)
```

```julia
savefig("Curves.pdf")
```
